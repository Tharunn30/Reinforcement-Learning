```python
def value_iteration_gamblers_problem(target, p_h=0.4, theta=0.0001):
    V = [0] * (target + 1)
    pi = [0] * (target + 1)
    delta = float('inf')

    while delta > theta:
        delta = 0
        for s in range(1, target):
            v = V[s]
            max_val = 0
            for a in range(1, min(s, target - s) + 1):
                win_state = s + a
                reward = 1 if win_state == target else 0
                val = p_h * (reward + V[win_state]) + (1 - p_h) * V[s - a]
                if val > max_val:
                    max_val = val
                    pi[s] = a
            V[s] = max_val
            delta = max(delta, abs(v - V[s]))

    return V, pi

target = 10
V, pi = value_iteration_gamblers_problem(target)
print("Optimal value function:")
print(V)
print("Optimal policy:")
print(pi)
```
